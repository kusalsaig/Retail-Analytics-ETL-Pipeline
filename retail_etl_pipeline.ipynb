```python
import pandas as pd
df = pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1')
print("Raw data shape:", df.shape)
df.head()
```

```python
# 🧹 Step 2: Clean Data
# Drop rows with missing CustomerID or Description
df.dropna(subset=['CustomerID', 'Description'], inplace=True)

# Remove negative quantities (usually returns)
df = df[df['Quantity'] > 0]

# Remove rows with UnitPrice <= 0
df = df[df['UnitPrice'] > 0]

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 🧾 Add TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

print("Cleaned data shape:", df.shape)
df.head()
```

```python
# 📊 Step 3: Example Analysis (Transform)
# Total revenue per country
revenue_by_country = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)
print("Top 5 Countries by Revenue:")
print(revenue_by_country.head())
```
